```python
import numpy as np
import pandas as pd

def combinations(n, k):
    return int(np.math.factorial(n) / (np.math.factorial(k) * np.math.factorial(n - k)))
```

## Задание 1

Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.

```python
a = np.array([100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150])
b = pd.DataFrame([100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150]) # для проверки
```

```python
M = a.sum()/len(a) # среднее арифметическое / математическое ожидание
m = b.mean() # проверка

M, m
```

```python
S = np.sqrt(((a - M)**2).sum() / len(a)) # среднее квадратичное отклонение
s = b.std(ddof=0) # проверка

S, s
```

```python
V1 = ((a - M)**2).sum() / len(a) # смещенная оценки дисперсий для данной выборки
V2 = ((a - M)**2).sum() / (len(a)-1) # несмещенная оценки дисперсий для данной выборки

v1 = b.var(ddof=0) # проверка
v2 = b.var(ddof=1) # проверка

V1, v1, V2, v2
```

## Задание 2

В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?

```python
c1w2 = combinations(5, 2) # 2 белых мяча из 1 ящика
c1w1 = combinations(5, 1) # 1 белый мяч из 1 ящика
c1o2 = combinations(3, 2) # 2 небелых мяча из 1 ящика
c1o1 = combinations(3, 1) # 1 небелый мяча из 1 ящика

c2w3 = combinations(5, 3) # 3 белых мяча из 1 ящика
c2w2 = combinations(5, 2) # 2 белых мяча из 1 ящика
c2w1 = combinations(5, 1) # 1 белый мяч из 1 ящика
c2o3 = combinations(6, 3) # 2 небелых мяча из 1 ящика
c2o2 = combinations(6, 2) # 2 небелых мяча из 1 ящика
c2o1 = combinations(6, 1) # 1 небелый мяча из 1 ящика

c1 = combinations(8, 2) # 2 мяча из 1 ящика
c2 = combinations(12, 4) # 4 мяча из 2 ящика

a1 = c1w2/c1 * c2w1*c2o3/c2 # достали 2 белых из 1 и 1 белый из 2го
a2 = c1w1*c1o1/c1 * c2w2*c2o2/c2 # достали 1 белый из 1 и 2 белых из 2го
a3 = c1o2/c1 * c2w3*c2o1/c2 # достали 0 белых из 1 и 3 белых из 2го

a = a1+a2+a3
print(f'Все 3 мяча белые {a}')
```

## Задание 3

На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен: a). первым спортсменом б). вторым спортсменом в). третьим спортсменом.

```python
# B - попадание в цель, Ai - выстрел произвёл i-й стрелок
# применим вормулу Байеса
A = 1/3 # вероятность того, что выстрелил i-й стрелок
P1 = A * 0.9
P2 = A * 0.8
P3 = A * 0.6

Pb = P1 + P2 + P3

P1 = A * P1 / Pb # попал 1й спортсмен
P2 = A * P2 / Pb # попал 2й спортсмен
P3 = A * P3 / Pb # попал 3й спортсмен

print(f'попал 1й спортсмен {P1}')
print(f'попал 2й спортсмен {P2}')
print(f'попал 3й спортсмен {P3}')
```

## Задание 4

В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же, сколько на A и B вместе. Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. Студент сдал первую сессию. Какова вероятность, что он учится: a). на факультете A б). на факультете B в). на факультете C?

Пусть на факультет A и B поступило по n человек. Тогда на факультет C поступило 2n человек. Всего у нас 4n стдентов.
Вероятность того, что студент из A сдал сессию PA = n/4n * 0.8
Вероятность того, что студент из B сдал сессию PB = n/4n * 0.7
Вероятность того, что студент из C сдал сессию PC = n/2n * 0.9

P = PA + PB + PC = n/4n * 0.8 + n/4n * 0.7 + n/2n * 0.9 - вероятность того, что студент сдал сессию

Вероятность того, что студент сдавший сессию из A:
PA * 0.8 / P

Вероятность того, что студент сдавший сессию из B:
PB * 0.7 / P

Вероятность того, что студент сдавший сессию из C:
PC * 0.9 / P

## Задание 5

Устройство состоит из трех деталей. Для первой детали вероятность выйти из строя в первый месяц равна 0.1, для второй - 0.2, для третьей - 0.25. Какова вероятность того, что в первый месяц выйдут из строя: а). все детали б). только две детали в). хотя бы одна деталь г). от одной до двух деталей?

```python
p1 = 0.1 # вышла из строя 1 деталь
q1 = 1 - 0.1

p2 = 0.2 # вышла из строя 2 деталь
q2 = 1 - 0.2

p3 = 0.25 # вышла из строя 3 деталь
q3 = 1 - 0.25

Pa = p1*p2*p3 # вышли из строя все детали
Pb = p1*p2*q3 + p1*q2*p3 + q1*p2*p3 # вышли из строя только 2 детали
Pc = 1 - q1*q2*q3 # вышла из строя хотя бы одна деталь
Pd1 = Pb + (p1*q2*q3 + q1*p2*q3 + q1*q2*p3) # вышли из строя от одной до двух деталей
Pd2 = Pc - Pa # или так

print(f'вышли из строя все детали {Pa}')
print(f'вышли из строя только 2 детали {Pb}')
print(f'вышла из строя хотя бы одна деталь {Pc}')
print(f'вышли из строя от одной до двух деталей {Pd1}')
print(f'или другой расчёт {Pd2}')
```
